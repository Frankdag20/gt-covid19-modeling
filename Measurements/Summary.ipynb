```python
# import packages
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
from datetime import datetime
from github import Github
import os
```

```python
df_cdc_cum_death = pd.read_csv("./measurements_cum_death_cdc.csv",index_col=[0,1]).drop(index=["US"])
df_cdc_cum_death.reset_index(inplace=True)
df_cdc_cum_death.set_index(["model", "state"], inplace=True)
df_cdc_cum_death.drop(index=["IHME-CurveFit"],inplace=True)

df_cdc_inc_death = pd.read_csv("./measurements_inc_death_cdc.csv",index_col=[0,1]).drop(index=["US"])
df_cdc_inc_death.reset_index(inplace=True)
df_cdc_inc_death.set_index(["model", "state"], inplace=True)
df_cdc_inc_death.drop(index=["IHME-CurveFit"],inplace=True)

df_cdc_inc_case = pd.read_csv("./measurements_inc_case_cdc.csv",index_col=[0,1]).drop(index=["US"])
df_cdc_inc_case.reset_index(inplace=True)
df_cdc_inc_case.set_index(["model", "state"], inplace=True)
df_cdc_inc_case.drop(index=["IHME-CurveFit", "COVIDhub-baseline"],inplace=True)

df_git_cum_death = pd.read_csv("./measurements_cum_death_GITIDEAS.csv",index_col=[0,1]).drop(index=["US"])
df_git_cum_death.reset_index(inplace=True)
df_git_cum_death.set_index(["model", "state"], inplace=True)

df_git_cum_case = pd.read_csv("./measurements_cum_case_GITIDEAS.csv",index_col=[0,1]).drop(index=["US"])
df_git_cum_case.reset_index(inplace=True)
df_git_cum_case.set_index(["model", "state"], inplace=True)

df_git_inc_case = pd.read_csv("./measurements_inc_case_GITIDEAS.csv",index_col=[0,1]).drop(index=["US"])
df_git_cum_case.reset_index(inplace=True)
df_git_cum_case.set_index(["model", "state"], inplace=True)

df_git_inc_death = pd.read_csv("./measurements_inc_death_GITIDEAS.csv",index_col=[0,1]).drop(index=["US"])
df_git_cum_case.reset_index(inplace=True)
df_git_cum_case.set_index(["model", "state"], inplace=True)
```

```python
print("---------------------------- Model Evaluation Summary (State-Level) ---------------------------------")
print("------------------- Models from CDC COVID forecast Github (State-Level) -----------------------------")
print("................................... Cumulative Deaths ................................................")
summary = pd.DataFrame(index=["MSE", "MAE", "MAPE"], columns=["Max", "Max_model", "Min", "Min_model", "Mean"])
temp = df_cdc_cum_death.groupby(["model"]).mean()
for i in ["MSE", "MAE", "MAPE"]:
    summary.loc[i]["Max"] = temp.max()[i]
    summary.loc[i]["Max_model"] = temp.idxmax()[i]
    summary.loc[i]["Min"] = temp.min()[i]
    summary.loc[i]["Min_model"] = temp.idxmin()[i]
    summary.loc[i]["Mean"] = temp.mean()[i]
print(summary)

print("................................... Incident Deaths ................................................")
temp = df_cdc_inc_death.groupby(["model"]).mean()
for i in ["MSE", "MAE", "MAPE"]:
    summary.loc[i]["Max"] = temp.max()[i]
    summary.loc[i]["Max_model"] = temp.idxmax()[i]
    summary.loc[i]["Min"] = temp.min()[i]
    summary.loc[i]["Min_model"] = temp.idxmin()[i]
    summary.loc[i]["Mean"] = temp.mean()[i]
print(summary)

print("................................... Incident Cases .................................................")
temp = df_cdc_inc_case.groupby(["model"]).mean()
for i in ["MSE", "MAE", "MAPE"]:
    summary.loc[i]["Max"] = temp.max()[i]
    summary.loc[i]["Max_model"] = temp.idxmax()[i]
    summary.loc[i]["Min"] = temp.min()[i]
    summary.loc[i]["Min_model"] = temp.idxmin()[i]
    summary.loc[i]["Mean"] = temp.mean()[i]
print(summary)

```
